```python
#Loading the data
df2 = pd.read_csv('Input2.csv')
df2.head()
```

```python
new_df1 = df1.iloc[: , [0, 2,3]].copy()
new_df1.head()
```

```python
new_df2 = df2.drop(["Milli_Sec"], axis = 1)
new_df2.head()
```

```python
#defining the function to remove punctuation
comment=pd.Series(index=new_df1['Comment'])
comment.head()
```

```python
new_df1['Unit']=comment.index.str.extract(r'\((\w+)\)', expand=True)
```

```python
new_df1.head()
```

```python
new_df1.shape
```

```python
new_df2.insert(2, "Time_Stamp", new_df2['Date'].astype(str) +" "+ new_df2["Time"], True)
new_df2=new_df2.drop(columns=['Date','Time'],axis=1)
```

```python
new_df2.head()
```

```python
new_df2.shape
```

```python
new_df2 =new_df2.groupby('Time_Stamp', as_index=False).max()
new_df2.head(10)
```

```python
Output = pd.DataFrame()
d = dict()

d['event_timestamp'] = []
d["tag__id"] = []
d["tag__name"] = []
d["tag__desc"] = []
d["tag__value"] = []
d["tag__unit"] = []

a2 = []
for k in d:
    a2.append(k)

alias = ["ItemId", 'ItemId', 'ItemName', 'Comment', 'ItemId','Unit']
for i in range(7):
    for j in range(6):
        d[a2[j]].append(new_df1[alias[j]][i])
    d[a2[0]][i] = new_df2["Time_Stamp"][i*10]
    d[a2[4]][i] = -212340981234091283470112342
    for j in range(i*10, min(i*10+10, 64)):
        d['tag__value'][i] = max(new_df2[new_df1[alias[2]][i]][j], d['tag__value'][i])

Output = pd.DataFrame(d)
Output
```

```python
Output.to_csv('Output.csv', index=False, na_rep='')
```
